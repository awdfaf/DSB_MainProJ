```python
voice_list = []
class voice:
    def __init__(self):
        self.r = sr.Recognizer()       
        self.text = ''      # 음성에서 변환된 텍스트
        self.token_dict = {} # 단어:횟수 딕셔너리 생성
        
    def call(self):
        for i in range(8, 29) :
            AUDIO_FILE = "D:/1조프로젝트/Voice_Phishing/피싱파일/type1/대출사기{}.wav".format(i)
            print('\n※ 통화 시작 ※\n')
            while True:
                with sr.AudioFile(AUDIO_FILE) as source:
                    voice = self.r.record(source)
                    self.text = self.r.recognize_google(voice, language='ko-KR')                    
                    print('▶ 통화내역 : {}'.format(self.text))
                    voice_list.append(self.text)
                break

            
```
